```python
# Replace salary 0 and NaN with median salary of 'Placed' students
median_salary = dataset[dataset['salary'] > 0]['salary'].median()
dataset['salary'] = dataset['salary'].replace(0, np.nan)
dataset['salary'].fillna(median_salary, inplace=True)
```
