```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import ttest_rel
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
not_placed_count = dataset[dataset['status'] == 'Not Placed'].shape[0]
print("Not Placed:", not_placed_count)
```

```python
#dataset[dataset['status'] == 'Not Placed'].describe()
# Step 5: Count not placed
not_placed = dataset[dataset['status'] == 'Not Placed']
print("Number of students not placed:", len(not_placed))
```

```python
# Step 6: Compare features of placed vs not placed
dataset.groupby('status')[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']].mean()
```

```python
# Step 7: Correlation
correlation = dataset[['salary', 'mba_p']].corr()
print(correlation)
```

```python
# Correlation
print("Correlation:", dataset['salary'].corr(dataset['mba_p']))

# Scatter plot
sns.scatterplot(x='mba_p', y='salary', data=dataset)
plt.title('Salary vs MBA Percentage')
plt.xlabel('MBA Percentage')
plt.ylabel('Salary')
plt.show()
```

```python
dataset.groupby('specialisation')['salary'].mean().sort_values()
```

```python
# Step 9: Count of students earning >500000
count_high_salary = dataset[dataset['salary'] > 500000].shape[0]
print("Number of students earning above 500000:", count_high_salary)
```

```python
anova_result = stats.f_oneway(dataset['etest_p'], dataset['mba_p'])
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

```python
from scipy.stats import ttest_ind

group1 = dataset[(dataset['degree_t'] == 'Sci&Tech')]['salary']
group2 = dataset[(dataset['specialisation'] == 'Mkt&HR')]['salary']

t_stat, p_val = ttest_ind(group1, group2, equal_var=False)
print("T-test p-value:", p_val)
```

```python
#"There is no significant difference between the means of the two groups as indicated by the t-test statistic of t=0.011
```

```python

def stdNBgraph(dataset):
    
        import seaborn as sns
        mean=dataset.mean()
        std=dataset.std()

        values=[i for i in dataset]

        z_score=[((j-mean)/std) for j in values]

        sns.distplot(z_score,kde=True)

        sum(z_score)/len(z_score)
    #z_score.std()
```

```python
stdNBgraph(dataset["salary"])
```

```python
from scipy.stats import norm
mean = dataset['salary'].mean()
std = dataset['salary'].std()
prob_density = norm.cdf(900000, mean, std) - norm.cdf(700000, mean, std)
print("Probability Density:", prob_density)
```

```python
sci_df = dataset[dataset['degree_t'] == 'Sci&Tech']
t_stat, p_val = ttest_ind(sci_df['etest_p'], sci_df['mba_p'], equal_var=False)
print("T-test p-value:", p_val)
```

```python
#"reject" the alternate hypotesis because the p>0.05
```
